```python
from numpy import random
def load_data(negative, positive):
    returnval = []
    with open(negative) as file1:
        content2 = file1.read().replace("\n", " ").split(">")
        content = [a.split(" ", 1) for a in content2]
        for a in content[1:]:
            if a != "" and a!= [""] and a != []:
                returnval.append([a[0], a[1].replace(" ", ""), 0])
    with open(positive) as file2:
        content2 = file2.read().replace("\n", " ").split(">")
        content = [a.split(" ", 1) for a in content2][:1000]
        for a in content[1:]:
            if a != "" and a != [""] and a != []:
                returnval.append([a[0], a[1].replace(" ", ""), 1])
    #random.shuffle(returnval,random.random)
    return returnval
```

```python
dataset = load_data("dataSet/protein_neg.txt","dataSet/protein_pos.txt")
```

```python
def get_base_frequencies(seq):
    return [ seq.count(base)/len(seq)
        for base in 'ATGC']
```

```python
with open("base_feature/raw_data_all.csv","r+") as csvfile:
    reader = csv.reader(csvfile)
    for line in reader:
        frequencies = get_base_frequencies(line[1])
        with open("base_feature/feature_ATGC_freq.csv","a+",newline='') as file:
            writer = csv.writer(file)
            writer.writerow(frequencies)
```

```python
import codecs
from itertools import *
import numpy as np
```

```python
def load_data(filename):
    file = codecs.open(filename,'r','utf-8')
    data = []
    label = []
    for line in islice(file,0,None):
        line = line.strip().split(',')
        #print ("reading data....")
        data.append([float(i) for i in line[0:-1]])
        label.append(line[-1])
    x = np.array(data)
    y = np.array(label)
    print (x)
    print (y)
    return x,y
```

```python
import pylab as pl
from itertools import *
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
def train(x_train,y_label):  ####为保证分类结果的准确度可靠，采用十折交叉验证
    ##########logisticRegression################
    clf1 = LogisticRegression()
    score1 = model_selection.cross_val_score(clf1,x_train,y_label,cv=10,scoring="accuracy")
    x = [int(i) for i in range(1,11)]
    y = score1
    pl.ylabel(u'Accuracy')
    pl.xlabel(u'times')
    pl.plot(x,y,label='LR')
    pl.legend()
    pl.savefig("1_LR.png")
    #pl.show()
    print ('The accuracy of LogisticRegression:')
    print (np.mean(score1))
    ###############SVM（linear)###################
    clf2 = svm.LinearSVC(random_state=2016)
    score2 = model_selection.cross_val_score(clf2,x_train,y_label,cv=10,scoring='accuracy')
    #print score2
    print ('The accuracy of linearSVM:')
    print ((np.mean(score2)))
    x = [int(i) for i in range(1, 11)]
    y = score2
    pl.ylabel(u'Accuracy')
    pl.xlabel(u'times')
    pl.plot(x, y,label='SVM')
    pl.legend()
    pl.savefig("2_SVM.png")
    #pl.show()
    #################Naive Bayes################
    clf3 = GaussianNB()
    score3 =  model_selection.cross_val_score(clf3,x_train,y_label,cv=10,scoring='accuracy')
    print ("The accuracy of Naive Bayes:")
    print ((np.mean(score3)))
    x = [int(i) for i in range(1, 11)]
    y = score3
    pl.ylabel(u'Accuracy')
    pl.xlabel(u'times')
    pl.plot(x, y,label='NB')
    pl.legend()
    pl.savefig("3_NB.png")
    #pl.show()
    ################DecidionTree###############
    clf4 = tree.DecisionTreeClassifier()
    score4 = model_selection.cross_val_score(clf4,x_train,y_label,cv=10,scoring="accuracy")
    print ('The accuracy of DB:')
    print (np.mean(score4))
    x = [int(i) for i in range(1, 11)]
    y = score4
    pl.ylabel(u'Accuracy')
    pl.xlabel(u'times')
    pl.plot(x, y,label='DB')
    pl.legend()
    pl.savefig("4_DB.png")
    #pl.show()
```

```python
X,Y = load_data('base_feature/feature_ATGC_freq.csv')
```

```python
train(X,Y)
```
